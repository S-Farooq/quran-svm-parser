```python
from bs4 import BeautifulSoup
import requests

def get_meanings(root):
    page = requests.get("https://corpus.quran.com/qurandictionary.jsp?q={}".format(root))
    if page.status_code==200:
        soup = BeautifulSoup(page.content, 'html.parser')
        
    # root_info = soup.find_all(["p"], {"class": ["dsm"]})[0]
    # also_info = soup.find_all(["ul"], {"class": ["also"]})[0]
    meanings = soup.find_all(["h4"], {"class": ["dxe"]})
    meanings = [x.text for x in meanings if "Verb (form" in x.text]

    return {"meanings":meanings}

def get_webpage_emd(loc):
    page = requests.get("https://corpus.quran.com/wordmorphology.jsp?location={}".format(loc))
    if page.status_code==200:
        soup = BeautifulSoup(page.content, 'html.parser')

    # print(soup.prettify())
    tokenCell = soup.find_all("td", {"class": "tokenCell"})
    # print(tokenCell)
    # morphologyCell = soup.find_all("td", {"class": "morphologyCell"})
    # grammarCell = soup.find_all("td", {"class": "grammarCell"})

    # desc = soup.find_all("form")[0].find_all("p")[0]

    # verse_trans = soup.find_all("p",{"class": "verseImage"})[0].find_next_sibling("p").text
    # print(tokenCell[0])
    # print(tokenCell[0].table.previous_sibling.text)
    meaning = tokenCell[0].table.previous_sibling.text.replace("\n","").strip()
    if meaning=="":
        raise("error getting meaning")
    return {
        # "word_image_src":tokenCell[0].find("a", {"class":"tokenLink"}).img["src"],
        # "desc":str(desc),
        "meaning": meaning,
        # "tokenCell": str(tokenCell),
           }

get_webpage_emd("2:8:6")
```

```python
import re 
EXCLUDE_LIST = []
# translation by verse, but indexed
by_verse_df = q_token_meaning
by_verse_df = by_verse_df.groupby(['chapter_number','verse_number']).agg(list).reset_index()
by_verse_df = by_verse_df.reset_index()

# by_verse_df = by_verse_df[by_verse_df["chapter_number"].isin([1,2])]

def flatten(xss):
    return [x for xs in xss for x in xs]


by_verse_df["sahih_translation_nlp"] = by_verse_df["translation"].apply(
    lambda c: [nlp(x) for x in c]
)

by_verse_df["sahih_translation_tokens"] = by_verse_df["sahih_translation_nlp"].apply(
    lambda c: [[str(x.text) for x in v if x.pos_ in elig_pos_tags.keys()] for v in c]
)

def get_translation_token_to_arb_token_map(tokens_list):
    map = []
    
    for i in range(len(tokens_list)):
        map.extend([i]*len(tokens_list[i]))
    return map

by_verse_df["sahih_translation_token_map"] = by_verse_df["sahih_translation_tokens"].apply(lambda c: get_translation_token_to_arb_token_map(c))
by_verse_df["sahih_translation_tokens"] = by_verse_df["sahih_translation_tokens"].apply(lambda c: flatten(c))
by_verse_df["translation_flat"] = by_verse_df["translation"].apply(lambda c: " ".join(c))

by_verse_df["sahih_translation_nlp"] = by_verse_df["translation_flat"].apply(
    lambda x: [t for t in nlp(x) if t.pos_ in elig_pos_tags.keys()]
)

def separate_by_map_map(tokens_list, map, index):
    try:
        assert(len(tokens_list) == len(map))
    except Exception as e:
        # print(tokens_list, map)
        # print(e)
        EXCLUDE_LIST.append(index)
        return []
        
    new_list = []
    for i in range(max(map)+1):
        new_list.append([])
        
    for i in range(len(tokens_list)):
        new_list[map[i]].append(tokens_list[i])
    return new_list

def add_to_eng_lemma_ser(lemma, LemSerEngIns):
    LemSerEngIns.add(lemma)
    return
    
by_verse_df["sahih_translation_nlp"] = by_verse_df.apply(
    lambda r: separate_by_map_map(r["sahih_translation_nlp"], r["sahih_translation_token_map"], r["index"]), axis=1
)

by_verse_df = by_verse_df[~by_verse_df.index.isin(EXCLUDE_LIST)]

LemSerEng = LemmaService()

by_verse_df["sahih_translation_pos"] = by_verse_df["sahih_translation_nlp"].apply(
    lambda c: [[(x.pos_) for x in v if x.pos_ in elig_pos_tags.keys()] for v in c]
)

by_verse_df["sahih_translation_lemma"] = by_verse_df["sahih_translation_nlp"].apply(
    lambda c: [[(x.lemma_) for x in v if x.pos_ in elig_pos_tags.keys()] for v in c]
)

by_verse_df["sahih_translation_lemma"].apply(
    lambda c: [[add_to_eng_lemma_ser(x, LemSerEng) for x in v] for v in c]
)

by_verse_df["sahih_translation_tag"] = by_verse_df["sahih_translation_nlp"].apply(
    lambda c: [[(x.tag_) for x in v if x.pos_ in elig_pos_tags.keys()] for v in c]
)

by_verse_df["sahih_translation_dep"] = by_verse_df["sahih_translation_nlp"].apply(
    lambda c: [[(x.dep_) for x in v if x.pos_ in elig_pos_tags.keys()] for v in c]
)

by_verse_df["sahih_translation_dep_headpos"] = by_verse_df["sahih_translation_nlp"].apply(
    lambda c: [[(x.head.pos_) for x in v if x.pos_ in elig_pos_tags.keys()] for v in c]
)


by_verse_df.shape, len(EXCLUDE_LIST), LemSerEng.count
```

```python
from scipy.sparse import lil_matrix
from random import sample
class MLProblem:

    def __init__(self):
        self._feature_vectors: List[List[int]] = []
        self._feature_count: int = 0
        self._labels: List[int] = []
        self._pos: List[int] = []
        self._neg: List[int] = []

    def add(self, instance: Instance, label):
        feature_vector = instance.feature_vector
        self._feature_vectors.append(feature_vector)
        if self._feature_count>0:
            assert(instance.size==self._feature_count)
        else:
            self._feature_count = instance.size
        self._labels.append(label)
        if label==1:
            self._pos.append(len(self._labels)-1)
        else:
            self._neg.append(len(self._labels)-1)

    def _neg_sample_indexes(self, n):
        return sample(self._neg,n)
        
    def _build_matrix_from_fv(self, feature_vectors):
        matrix = lil_matrix((len(feature_vectors), self._feature_count))
            
        for i, fv in enumerate(feature_vectors):
            for j in fv:
                matrix[i, j] = 1
        return matrix
        
    def build_matrix(self, pos_only=False, neg_only=False, stratified_sampling=True, n_sample=None):
        
        all_labels = []
        
        if pos_only:
            idxs = self._pos
        elif neg_only:
            idxs = self._neg
        elif stratified_sampling:
            idxs = self._pos + self._neg_sample_indexes(len(self._pos))
        else:
            idxs = list(range(len(self._labels)))

        if n_sample and n_sample>1:
            idxs = sample(idxs,n_sample)
        elif n_sample and n_sample<1:
            idxs = sample(idxs,int(n_sample*len(idxs)))
            
        matrix = self._build_matrix_from_fv(
            [self._feature_vectors[idx] for idx in idxs]
        )
        all_labels = [self._labels[idx] for idx in idxs]


        return (matrix, all_labels)

ml_problem = MLProblem()
for i in range(len(instance_objs)):
    ml_problem.add(instance_objs[i], ml_labels[i])
    
```
